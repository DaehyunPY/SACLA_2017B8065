```python
import pickle
import os
import textwrap

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from lmfit.models import GaussianModel
import altair as alt
from altair import datum

from dltools.cov.markup import compute

alt.renderers.enable('notebook')
```

```python
filename = "Data/Cov ang dist and KER at the low int group; target=C2p,I4p,I3p & h0phi=-135--120.pickle"
```

```python
def binit(x):
    return x.groupby_bins("ke", range(0, 200, 6), right=False)


with open(filename, "rb") as fp:
    d = pickle.load(fp)

xkey, ykey = "ke", "sumke"
n = d["N"] * 360 / 255
dx = d["Sum[XYZ]"][xkey][1] - d["Sum[XYZ]"][xkey][0]
x = binit(d["Sum[XYZ]"][xkey] + dx / 2).mean()
y = binit(d["Sum[XYZ]"][ykey]).sum()
# where = 0 < y
where = slice(None)

df = pd.concat([
    pd.DataFrame({
        "Total KE (eV)": x[where],
        "Type": "Cov",
        "Yield (count)": binit(d["Cov[X,Y,Z]"][ykey] * n).sum()[where],
        "Std (count)": binit((d["Err[Cov[X,Y,Z]]"][ykey] * n) ** 2).sum()[where] ** 0.5,
    }),   
], ignore_index=True)

base = alt.Chart(df, width=80, height=80).encode(
    alt.X(
        "Total KE (eV)",
        scale=alt.Scale(domain=[50, 200], nice=False),
        axis=alt.Axis(tickCount=4)
    ),
    y="Yield (count)",
    yError="Std (count)",
)
(
    base.mark_errorbar(clip=True, color="#F58518")
    + base.mark_point(clip=True, filled=True, size=4, opacity=1, color="#F58518")
)
```
