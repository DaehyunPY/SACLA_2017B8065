```python
import pickle
import os
import textwrap

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from lmfit.models import GaussianModel

from dltools.cov.markup import compute
```

```python
filename = "Data/Cov ang dist and KER at the low int group; target=C1p,I1p,I1p & h1phi=135--225.pickle"
```

```python
with open(filename, "rb") as fp:
    d = pickle.load(fp)

xkey, ykey = "ke", "sumke"
n = d["N"]
dx = d["Cov[X,Y,Z]"][xkey][1] - d["Cov[X,Y,Z]"][xkey][0]
x = d["Cov[X,Y,Z]"][xkey] + dx / 2
where = 0 < d["Sum[XYZ]"][ykey]

plt.figure()
plt.xlabel("KER")
plt.ylabel("Count")
plt.errorbar(
    x[where],
    d["Sum[XYZ]"][ykey][where],
    d["Sum[XYZ]"][ykey][where] ** 0.5,
    fmt=" ",
    capsize=2,
    label="Cov",
)
plt.errorbar(
    x[where],
    (d["Sum[XYZ]"][ykey] - d["Cov[X,Y,Z]"][ykey] * n)[where],
    ((d["Err[Cov[X,Y,Z]]"][ykey] * n) ** 2 - d["Sum[XYZ]"][ykey])[where] ** 0.5,
    fmt=" ",
    capsize=2,
    label="Cov",
)
plt.errorbar(
    x[where],
    d["Cov[X,Y,Z]"][ykey][where] * n,
    d["Err[Cov[X,Y,Z]]"][ykey][where] * n,
    fmt=" ",
    capsize=2,
    label="Cov",
)
plt.minorticks_on()
plt.grid(True, which="both")
plt.xlim(0, 150)
plt.tight_layout()
plt.show()
```
