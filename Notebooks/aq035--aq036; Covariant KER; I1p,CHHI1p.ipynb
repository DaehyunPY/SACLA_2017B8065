```python
# %% Load PySpark
builder = (
    SparkSession
    .builder
    .config("spark.executor.memory", "16g")
    .config("spark.driver.memory", "16g")
    .config("spark.driver.maxResultSize", "4g")
    .config(        
        "spark.jars.packages",
        "org.apache.hadoop:hadoop-aws:2.7.0,"
        "org.mongodb.spark:mongo-spark-connector_2.11:2.3.1,"
        "org.diana-hep:spark-root_2.11:0.1.15,"
    )
)

print("Loading PySpark...")
spark = builder.getOrCreate()
print("Done!")
```

```python
# %% Load data
from pyspark.sql.types import (
    ArrayType, BooleanType, LongType, StructField, StructType, DoubleType,
)
from dltools import SpkHits


print("Loading data...")
df = (
    spark
    .read
    .format("com.mongodb.spark.sql.DefaultSource")
    .option("uri", "mongodb://mongodb/sacla_2017b8065.resorted")
    .option("pipeline", """[
        {
            $match: {
                aq: {$in: [35, 36]},
            },
        },
        {
            $project: {
                tag: true,
                hits: {
                    $filter: {
                        input: "$hits",
                        as: "item",
                        cond: {$lte: ["$$item.flag", 6]},
                    },
                },
            },
        },
    ]""")
    .schema(
        StructType([
            StructField("tag", LongType()),
            StructField("hits", SpkHits)
        ])
    )
    .load()
)
df.printSchema()
print("Done!")
```

```python
import typing
import pyspark
from cov import cov2d_complicated


def ishit0(hit: pyspark.sql.Row) -> bool:
    return "I1p" in hit["as_"]


def ishit1(hit: pyspark.sql.Row) -> bool:
    return "CHHI1p" in hit["as_"]


combined = cov2d_complicated(
    df, ishit0, ishit1,
    npart=60,
    opt1={
        "npart": 12,
    },
    opt2={
        "fraction": 0.1,
        "npart": 12,
    },
)
```

```python
def hist_ker(rows: typing.List[pyspark.sql.Row]) -> typing.List[np.ndarray]:
    ret, *_ = np.histogram(
        [
            r["hit0"]["as_"]["I1p"]["ke"]
            + r["hit1"]["as_"]["CHHI1p"]["ke"]
            for r in rows
        ],
        bins=np.linspace(0, 300, 601),
    )
    return [ret]
```

```python
x = np.linspace(0, 300, 601)[1:]
n = combined(hist_ker)["N"]

plt.figure(figsize=(6, 6))
plt.step(
    x, combined(hist_ker)["Cov[X,Y]"] * n,
    label="Cov[X,Y] * N",
)
plt.step(
    x, combined(hist_ker)["Sum[XY]"],
    label="Sum[XY]",
)
plt.step(
    x, combined(hist_ker)["Sum[X]Sum[Y]"] / n,
    label="Sum[X]Sum[Y] / N",
)
plt.xlim(0, 60)
plt.ylim(0, None)
plt.minorticks_on()
plt.grid(True, which="both")
plt.legend(shadow=True)

plt.tight_layout()
# plt.savefig("Figures/aq035--aq036; Covariant KER; I1p and CHHI1p.pdf")
plt.show()
```

```python
def hist_pyz(rows: typing.List[pyspark.sql.Row]) -> typing.List[np.ndarray]:
    data = np.array([
        [r["hit0"]["as_"]["I1p"]["py"] + r["hit1"]["as_"]["CHHI1p"]["py"],
         r["hit0"]["as_"]["I1p"]["pz"] + r["hit1"]["as_"]["CHHI1p"]["pz"]]
        for r in rows
    ]).reshape(-1, 2)
    ret, *_ = np.histogram2d(
        data[:, 0],
        data[:, 1],
        bins=(
            np.linspace(-200, 200, 201),
            np.linspace(-200, 200, 201),
        ),
    )
    return [ret]
```

```python
x = np.linspace(-200, 200, 201)[1:]
n = combined(hist_ker)["N"]

plt.figure(figsize=(6, 6))
plt.subplot(211)
plt.text(1, 1, "Total momentum of y-axis",
         horizontalalignment="right",
         verticalalignment="top",
         transform=plt.gca().transAxes,
         bbox={"facecolor": "w", "edgecolor": "w"})
ax = 1
plt.step(
    x, combined(hist_pyz)["Cov[X,Y]"].sum(ax) * n,
    label="Cov[X,Y] * N",
)
plt.step(
    x, combined(hist_pyz)["Sum[XY]"].sum(ax),
    label="Sum[XY]",
)
plt.step(
    x, combined(hist_pyz)["Sum[X]Sum[Y]"].sum(ax) / n,
    label="Sum[X]Sum[Y] / N",
)
plt.minorticks_on()
plt.grid(True, which="both")
plt.legend(loc="upper left", shadow=True)
plt.gca().label_outer()

plt.subplot(212)
plt.text(1, 1, "Total momentum of z-axis",
         horizontalalignment="right",
         verticalalignment="top",
         transform=plt.gca().transAxes,
         bbox={"facecolor": "w", "edgecolor": "w"})
ax = 0
plt.step(
    x, combined(hist_pyz)["Cov[X,Y]"].sum(ax) * n,
    label="Cov[X,Y] * N",
)
plt.step(
    x, combined(hist_pyz)["Sum[XY]"].sum(ax),
    label="Sum[XY]",
)
plt.step(
    x, combined(hist_pyz)["Sum[X]Sum[Y]"].sum(ax) / n,
    label="Sum[X]Sum[Y] / N",
)
plt.minorticks_on()
plt.grid(True, which="both")
# plt.legend(shadow=True)
plt.gca().label_outer()

plt.tight_layout()
plt.savefig("Figures/aq035--aq036; Covariant total momentum; I1p and CHHI1p.pdf")
plt.show()
```
